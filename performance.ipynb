```python
def criar_df_value_counts_proporcao(df,coluna,lookup_table,nome_index,nome_coluna='Proporcao (%)'):
    x=pd.DataFrame(df[coluna].value_counts()/len(df))    #criar novo dataframe a partir do value_counts dividido pelo número total de rows    
    x.index=[lookup_table[i] for i in x.index.to_list()] #renomear os valores no index 
    x.index.names=[nome_index]                           #renomear o index                               
    x=x.rename(columns={'count':nome_coluna})            #renomear a coluna
    return x
```

### Age

```python
# Proporcao de alunos em cada grupo de idade

index_idade={1:'18-21',2:'22-25',3:'>=26'}
idade_higher_education=criar_df_value_counts_proporcao(df_higher_education,"AGE",index_idade,'Idade')
idade_higher_education
```

```python
#Agrupar alunos por idade e nota

df_he_grouped_age_grade=df_higher_education.groupby(["AGE","GRADE"]).count().loc[:,['STUDENTID']].rename(columns={'STUDENTID':'Número Alunos'})
print('Total de Alunos: {}'.format(len(df_higher_education)))
df_he_grouped_age_grade
```

```python
#Proporção em relação ao total de alunos
#Grupo de idade e nota

sns.countplot(data=df_higher_education,x='AGE',hue='GRADE',stat='proportion',palette='Blues')
plt.xticks(ticks=[0,1,2
],labels=[index_idade[i]for i in range(1,4)])
```

```python
#Proporção dentro de cada grupo
#Por exemplo, dentre alunos com idades entre 18 e 21 anos, aproximadamente 20% tiraram nota 1


fig,axs=plt.subplots(ncols=3,nrows=1,figsize=(22,6))
for age_group in range(1,4):
    age=df_he_grouped_age_grade.loc[age_group,:]/df_he_grouped_age_grade.loc[age_group,:].sum()
    sns.barplot(data=age,x=age.index,y='Número Alunos',ax=axs[age_group-1])
    axs[age_group-1].set_xlabel('Nota')
    axs[age_group-1].set_ylabel('Proporção')
    axs[age_group-1].set_title('Idade: {}'.format(index_idade[age_group]))
```
